```python
# Cell 1 — Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('../data/creditcard.csv')
print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nBasic Statistics:\n")
df.describe()
```

```python
# Cell 4 — Check Class Imbalance
print("Class Distribution:")
print(df['Class'].value_counts())
print("\nFraud Percentage:", round(df['Class'].sum() / len(df) * 100, 4), "%")

# Visualize
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Fraud vs Normal Transactions')
plt.xticks([0,1], ['Normal', 'Fraud'])
plt.show()
```

```python
# Cell 5 — Amount Analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Normal transactions
axes[0].hist(df[df['Class']==0]['Amount'], bins=50, color='blue', alpha=0.7)
axes[0].set_title('Normal Transaction Amounts')
axes[0].set_xlabel('Amount')
axes[0].set_ylabel('Count')

# Fraud transactions
axes[1].hist(df[df['Class']==1]['Amount'], bins=50, color='red', alpha=0.7)
axes[1].set_title('Fraud Transaction Amounts')
axes[1].set_xlabel('Amount')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()

print("Average Normal Transaction:", round(df[df['Class']==0]['Amount'].mean(), 2))
print("Average Fraud Transaction:", round(df[df['Class']==1]['Amount'].mean(), 2))
```

```python
# Cell 6 — Correlation Heatmap
plt.figure(figsize=(16, 10))
correlation = df.corr()
sns.heatmap(correlation, 
            cmap='coolwarm',
            center=0,
            fmt='.1f',
            linewidths=0.5)
plt.title('Correlation Heatmap of All Features')
plt.show()
```

```python
# Cell 7 — Which features correlate most with Fraud
correlation_with_fraud = df.corr()['Class'].sort_values()

plt.figure(figsize=(10, 6))
correlation_with_fraud.drop('Class').plot(kind='bar', 
                                          color=['red' if x < 0 else 'blue' 
                                                 for x in correlation_with_fraud.drop('Class')])
plt.title('Feature Correlation with Fraud (Class)')
plt.xlabel('Features')
plt.ylabel('Correlation Score')
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()

print("\nTop 5 features MOST correlated with fraud:")
print(correlation_with_fraud.drop('Class').abs().sort_values(ascending=False).head())
```

```python
# Cell 8 — Time Analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Normal transactions over time
axes[0].scatter(df[df['Class']==0]['Time'], 
                df[df['Class']==0]['Amount'], 
                alpha=0.1, color='blue', s=1)
axes[0].set_title('Normal Transactions Over Time')
axes[0].set_xlabel('Time (seconds)')
axes[0].set_ylabel('Amount')

# Fraud transactions over time
axes[1].scatter(df[df['Class']==1]['Time'], 
                df[df['Class']==1]['Amount'], 
                alpha=0.5, color='red', s=5)
axes[1].set_title('Fraud Transactions Over Time')
axes[1].set_xlabel('Time (seconds)')
axes[1].set_ylabel('Amount')

plt.tight_layout()
plt.show()
```

```python
# Cell 9 — Distribution of Top Fraud Features
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.flatten()

top_features = ['V17', 'V14', 'V12', 'V10', 'V16', 'Amount']

for i, feature in enumerate(top_features):
    axes[i].hist(df[df['Class']==0][feature], 
                 bins=50, alpha=0.5, 
                 color='blue', label='Normal', density=True)
    axes[i].hist(df[df['Class']==1][feature], 
                 bins=50, alpha=0.5, 
                 color='red', label='Fraud', density=True)
    axes[i].set_title(f'{feature} Distribution')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Density')
    axes[i].legend()

plt.suptitle('Top Features: Fraud vs Normal Distribution', fontsize=14)
plt.tight_layout()
plt.show()
```
